```python
import yfinance
import numpy
import matplotlib
import pandas
```

```python
ticker_symbol = "GC=F"
ticker = yfinance.Ticker(ticker_symbol)
historical_data = ticker.history(period="1y")
```

```python
df=pandas.DataFrame()
df['Close']=historical_data['Close']
```

```python
matplotlib.pyplot.plot(df)
```

```python
df['RETURNS'] = numpy.log(df['Close']/df['Close'].shift(1))
```

```python
mean_return=df['RETURNS'].mean()
std_return=df['RETURNS'].std()
sharpe_ratio_return=mean_return/std_return
```

```python
mean_return=df['RETURNS'].mean()
std=df['Close'].std()
sharpe_ratio_classic=mean_return/std
```

```python
mean_return=df['RETURNS'].mean()
ups_std_return=df[df['RETURNS'] > mean_return]['RETURNS'].std()
upsortino_ratio_return=mean_return/ups_std_return
```

```python
mean_return=df['RETURNS'].mean()
ds_std_return=df[df['RETURNS'] < mean_return]['RETURNS'].std()
dsortino_ratio_return=mean_return/ds_std_return
```

```python
df['rolling_sharpe_return']=df['RETURNS'].rolling(window=21).mean()/df['RETURNS'].rolling(window=21).std()
```

```python
matplotlib.pyplot.plot(df['rolling_sharpe_return'])
```

```python
df['rolling_sharpe_classic']=df['RETURNS'].rolling(window=21).mean()/df['Close'].rolling(window=21).std()
```

```python
matplotlib.pyplot.plot(df['rolling_sharpe_classic'])
```

```python
df['rolling_ups_return']=df['RETURNS'].rolling(window=21).mean()/df[df['RETURNS'] > 0]['RETURNS'].rolling(window=21).std()
```

```python
df['rolling_ds_return']=df['RETURNS'].rolling(window=21).mean()/df[df['RETURNS'] < 0]['RETURNS'].rolling(window=21).std()
```
