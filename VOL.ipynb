```python
import yfinance
import numpy
import matplotlib
import pandas
```

```python
ticker_symbol = "GC=F"
ticker = yfinance.Ticker(ticker_symbol)
historical_data = ticker.history(period="1y")
```

```python
df=pandas.DataFrame()
df['Open']=historical_data['Open']
df['High']=historical_data['High']
df['Low']=historical_data['Low']
df['Close']=historical_data['Close']
df['Volume']=historical_data['Volume']
```

```python
matplotlib.pyplot.plot(df)
```

```python
c_vol=df['Close'].std()
```

```python
df['rolling_c_vol']=df['Close'].rolling(window=21).std()
```

```python
matplotlib.pyplot.plot(df['rolling_c_vol'])
```

```python
p_vol=((numpy.log(df['High']/df['Low']))**2).mean()/(4*numpy.log(2))
```

```python
df['rolling_p_vol']=((numpy.log(df['High']/df['Low']))**2).rolling(window=21).mean()/(4*numpy.log(2))
```

```python
matplotlib.pyplot.plot(df['rolling_p_vol'])
```

```python
gk_vol=((numpy.log(df['High']/df['Low']))**2 - (4*numpy.log(2)-2)*(numpy.log(df['Close']/df['Open']))**2).mean()/2
```

```python
df['rolling_gk_vol']=((numpy.log(df['High']/df['Low']))**2 - (4*numpy.log(2)-2)*(numpy.log(df['Close']/df['Open']))**2).rolling(window=21).mean()/2
```

```python
matplotlib.pyplot.plot(df['rolling_gk_vol'])
```

```python
df['rolling_rs_vol']=((numpy.log(df['High']/df['Open']))*(numpy.log(df['High']/df['Close'])) +  (numpy.log(df['Low']/df['Open']))*(numpy.log(df['Low']/df['Close']))).rolling(window=21).mean()
```

```python
matplotlib.pyplot.plot(df['rolling_rs_vol'])
```
