```python
import numpy
import matplotlib
import pandas
```

```python
mean_return=0.001
std_return=0.1
```

```python
start=100
L=1000
mu=mean_return
sigma=std_return
process=[start]
for k in range(L):
  z=2*numpy.random.uniform(0,1)-1
  process.append(process[k]*numpy.exp(mu+ z*sigma))
```

```python
df_process=pandas.DataFrame()
df_process['process']=process
```

```python
matplotlib.pyplot.plot(df_process)
```

```python
df_process['process'].plot.hist(by=None, bins=20,color='g')
```

```python
count, division = numpy.histogram(df_process['process'],bins=10)
```

```python
count
```

```python
n=len(count)
indexes=[]
for k in range(1,n-1):
  if count[k]>count[k-1] and count[k]>count[k+1]:
    indexes.append(k)
```

```python
indexes
```

```python

for j in indexes:
  matplotlib.pyplot.plot(division[j]*numpy.ones(L),color='g')
matplotlib.pyplot.plot(df_process)
```
