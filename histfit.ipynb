```python
import yfinance
import numpy
import matplotlib
import pandas
```

```python
ticker_symbol = "GC=F"
ticker = yfinance.Ticker(ticker_symbol)
historical_data = ticker.history(period="1y")
```

```python
df=pandas.DataFrame()
df['Close']=historical_data['Close']
```

```python
df['RETURNS'] = numpy.log(df['Close']/df['Close'].shift(1))
```

```python
def N(mu,sigma):
  return lambda x: 1/(sigma*numpy.sqrt(2*numpy.pi))*numpy.exp(-(x-mu)**2/(2*sigma**2))
```

```python
import matplotlib.pyplot
n=N(df['RETURNS'].mean(),df['RETURNS'].std())
x=numpy.linspace(df['RETURNS'].min(),df['RETURNS'].max(),100)
matplotlib.pyplot.plot(x,n(x))
df['RETURNS'].plot.hist(by=None, bins=20,color='c')
```
