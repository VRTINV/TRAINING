```python
import yfinance
import numpy
import matplotlib
import pandas
```

```python
ticker_symbol = "GC=F"
ticker = yfinance.Ticker(ticker_symbol)
historical_data = ticker.history(period="1y")
```

```python
df=pandas.DataFrame()
df['Close']=historical_data['Close']
```

```python
df['RETURNS'] = numpy.log(df['Close']/df['Close'].shift(1))
```

```python
mean_return=df['RETURNS'].mean()
std_return=df['RETURNS'].std()
```

```python
L=len(df)-1
mu=mean_return
sigma=std_return
process=[df['Close'].iloc[0]]
for k in range(L):
  z=2*numpy.random.uniform(0,1)-1
  process.append(process[k]*numpy.exp(mu+ z*sigma))
```

```python
df_process=pandas.DataFrame()
df_process['process']=process
df_process.index=df.index
```

```python
matplotlib.pyplot.plot(df_process)
```
